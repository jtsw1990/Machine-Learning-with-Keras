```python
import pandas as pd
import numpy as np
import os
from keras.models import Sequential
from keras.layers import Dense, Activation
os.chdir('/Users/jtsw/Desktop/Coding File/ML Datasets')
```

```python
df = np.loadtxt('pima-indians-diabetes.data.txt',delimiter=',')
print(df.shape)
#there are 768 data points and 8 attributes and 1 target
#1. number of times pregnant
#2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
#3. Diastolic blood pressure (mm Hg)
#4. Triceps skin fold thickness (mm)
#5. 2-Hour serum insulin (mu U/ml)
#6. Body mass index (weight in kg/(height in m)^2)
#7. Diabetes pedigree function
#8. Age (years)
#9. Class variable (0 or 1)
```

# Start with some EDA

```python
df_format = pd.DataFrame(df)
df_format.columns = ['preg','gluc','bldprs','sft','insu','BMI','DPF','age','diab']
df_format.head()
```

# Begin Building Model

```python
inputs = df[:,0:8] #inputs being the predictors
outputs = df[:,8] #outputs being the diabetic status
print(inputs.shape)
print(outputs.shape)
```

```python
model = Sequential()
model.add(Dense(12, input_shape = (8,), activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(inputs, outputs, epochs = 150, batch_size = 10)
```

```python
scores = model.evaluate(inputs, outputs)
res = [0,1]
for i in res:
    if i == 0:
        print(str(model.metrics_names[i]) + ' is ' + str(scores[i]))
    else:
        print(str(model.metrics_names[i]) + ' is ' + str(scores[i]*100) + '%')
```
